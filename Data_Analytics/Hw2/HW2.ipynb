```python
from sklearn.linear_model import Lasso
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler

# Assume image_matrix is (400, 2577), with the last column as gender labels
X = image_matrix[:, :-1]  # Extract the first 2576 columns (pixel values)
y = image_matrix[:, -1]   # Extract the last column (gender labels)

# Standardize X for better Lasso performance
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 1: Use Lasso to select important features
lasso = Lasso(alpha=0.01)  # Alpha is the regularization strength (adjustable)
lasso.fit(X_scaled, y)

# Get selected features (non-zero coefficients)
selected_features = np.where(lasso.coef_ != 0)[0]  # Indices of important features
X_reduced = X_scaled[:, selected_features]  # Keep only selected features

# Step 2: Perform Stepwise Regression on Reduced Features
def stepwise_selection(X, y, threshold_in=0.05, threshold_out=0.1):
    """
    Perform stepwise regression (forward selection + backward elimination)
    on a reduced set of features selected by Lasso.
    """
    selected_features = []
    remaining_features = list(range(X.shape[1]))  # All feature indices
    current_score, best_new_score = 0.0, 0.0
    
    while remaining_features:
        pvals = []
        for feature in remaining_features:
            test_features = selected_features + [feature]
            X_test = sm.add_constant(X[:, test_features])  # Add intercept
            model = sm.OLS(y, X_test).fit()
            pvals.append((feature, model.pvalues[-1]))  # Get p-value of the new feature

        # Select feature with the lowest p-value (most significant)
        feature_to_add, pval = min(pvals, key=lambda x: x[1])

        if pval < threshold_in:
            selected_features.append(feature_to_add)
            remaining_features.remove(feature_to_add)
            current_score = sm.OLS(y, sm.add_constant(X[:, selected_features])).fit().rsquared
        else:
            break  # Stop if no more significant features
        
        # Backward elimination
        while selected_features:
            X_test = sm.add_constant(X[:, selected_features])
            model = sm.OLS(y, X_test).fit()
            pvals = model.pvalues[1:]  # Ignore intercept p-value
            worst_pval = max(pvals) if len(pvals) > 0 else 0

            if worst_pval > threshold_out:
                worst_feature = selected_features[pvals.argmax()]
                selected_features.remove(worst_feature)
            else:
                break  # Stop if all remaining features are significant

    return selected_features

# Perform stepwise regression on the reduced dataset
final_selected_features = stepwise_selection(X_reduced, y)

# Create a binary mask of selected pixels
pixel_map = np.zeros(2576)
pixel_map[selected_features[final_selected_features]] = 1  # Mark selected pixels

# Reshape into (46, 56) for visualization
pixel_image = pixel_map.reshape(46, 56)

# Plot the selected pixels
plt.figure(figsize=(6, 6))
plt.imshow(pixel_image, cmap="gray", aspect="auto")
plt.colorbar(label="Selected Pixel (1 = Important, 0 = Not Important)")
plt.title("Lasso + Stepwise Regression: Important Pixels for Gender Prediction")
plt.xlabel("Width (56 pixels)")
plt.ylabel("Height (46 pixels)")
plt.show()
```
